Automating selection of the different models

```python
# to fit all unique ARIMA models. By calculating AIC


def optimize_ARIMA(endog: Union[pd.DataFrame, list], order_list: list, d: int) -> pd.DataFrame:
    
    results = []
    
    for order in tqdm(order_list):
        try: 
            model = SARIMAX(endog, order=(order[0], d, order[1]), simple_differencing=False).fit(disp=False)
        except:
            continue
            
        aic = model.aic
        results.append([order, aic])
        
    result_df = pd.DataFrame(results)
    result_df.columns = ['(p,q)', 'AIC']
    
    #Sort in ascending order, lower AIC is better
    result_df = result_df.sort_values(by='AIC', ascending=True).reset_index(drop=True)
    
    return result_df
```

```python
#creating a list of all possible ps 
ps = range(0, 4, 1) 

#creating a list of all possible qs
qs = range(0, 4, 1)

#number of time series was differenced
d = 0

# generating a list containing all unique combinations
order_list = list(product(ps, qs))
```

```python
# creating our test and training data
# Data we will use to fit the model
train = df[0:]

# the last 3 values we are trying to predict
test = df[-3:]
```

Checking what data type train and test data are.

```python
def check_data_type(data):
    if isinstance(data, pd.Series):
        print("The data is a Pandas Series.")
    elif isinstance(data, pd.DataFrame):
        print("The data is a Pandas DataFrame.")
    elif isinstance(data, list):
        print("The data is a Python list.")
    else:
        print(f"The data is of type {type(data)}.")
```

```python
# we want to run the the function to obtain the model with the lowest AIC
# this will help us calculate how likely is it that our observed data is coming from a give selection of p,d,q that we pass through the ARIMA function

result_df = optimize_ARIMA(train, order_list, d)
result_df
```

```python
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mape_arima= mape(test['EXPORT'],test['ARIMA(1,0,2)'])
mape_arima1= mape(test['EXPORT'],test['ARIMA(1,0,1)'])


print(mape_arima,mape_arima1)
```
